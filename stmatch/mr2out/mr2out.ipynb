```python
import osmnx as ox
import networkx as nx
import pandas as pd
import matplotlib.pyplot as plt
import geopandas as gpd
from shapely.geometry import Point, LineString
```

```python
mr = pd.read_csv('mr.csv',sep=';')
mr
```

```python
mr_valid = mr[~mr['cpath'].isnull()]
mr_valid
```

```python
road = gpd.read_file('../road.csv')
road
```

```python
road['id']=road['id'].astype(int)
road['old_id'] = road['old_id'].astype(int)
```

```python
def func(row):
    tmp = []
    tmp = row['cpath'].split(',')
    tmp = [eval(item) for item in tmp]
    tmp = list(road.loc[road['id'].isin(tmp),'old_id'].unique())
    return tmp

mr_valid['old_id'] = mr_valid.apply(func, axis=1)
mr_valid
```

```python
mr_valid.to_csv('mr_valid.csv')
```
