```python
import osmnx as ox
import networkx as nx
import pandas as pd
import matplotlib.pyplot as plt
import geopandas as gpd
from shapely.geometry import Point, LineString
```

```python
road = pd.read_csv('data/road1.csv')
road
```

```python
node = pd.read_csv('node_mid.csv')
node
```

利用生成的node信息给road表加上target和source字段

```python
def func(row):
    for i in range(len(node)):
        if row['coordinate2_lon'] == node.loc[i, 'lon'] and row['coordinate2_lat'] == node.loc[i, 'lat']:
            return node.loc[i, 'id'] 
road['target'] = road.apply(func, axis=1)
road.to_csv('target_attached_road.csv')
```

```python
def func(row):
    for i in range(len(node)):
        if row['coordinate1_lon'] == node.loc[i, 'lon'] and row['coordinate1_lat'] == node.loc[i, 'lat']:
            return node.loc[i, 'id'] 
road['source'] = road.apply(func, axis=1)
road.to_csv('source_attached_road.csv')
```

```python
road_source = pd.read_csv('source_attached_road.csv')
road_target = pd.read_csv('target_attached_road.csv')
```
