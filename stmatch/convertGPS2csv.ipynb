```python
import pandas as pd
from shapely.geometry import Point, LineString
```

```python
traj = pd.read_csv('data/traj.csv')
traj
```

```python
df = pd.DataFrame(columns=['id', 'geom'])
df
```

```python
traj.iloc[1, 4]
```

```python
df = pd.DataFrame(columns=['id', 'geom'])
groupDf = traj.groupby('traj_id')
for name,group in groupDf:
    tmp = []
    for index in range(len(group)):
        row = eval(group.iloc[index, 4])
        tmp.append(tuple(row)) 
    df = pd.concat([df,pd.DataFrame({'id':[name], 'geom':[LineString(tmp)]})])
df
```

```python
df.to_csv('traj_LineString.csv',index=False)
```
