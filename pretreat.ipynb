```python
import pandas as pd
import ast
from geopy.distance import geodesic
from math import *
```

```python
road = pd.read_csv('data/road.csv')
road
```

```python
df = pd.DataFrame(columns=['old_id', 'coordinate1', 'coordinate2', 'coordinate1_lon','coordinate1_lat', 'coordinate2_lon', 'coordinate2_lat'])
def func(row):
    global df
    List = [tuple(sublist) for sublist in ast.literal_eval(row['coordinates'])]
    if len(List) > 2:
        for i in range(len(List)-1):
            new_row = pd.DataFrame({'old_id':row['id'], 'coordinate1':[List[i]], 'coordinate2':[List[i+1]], 'coordinate1_lon':[List[i][0]], 'coordinate1_lat':[List[i][1]], 'coordinate2_lon':[List[i+1][0]], 'coordinate2_lat':[List[i+1][1]]})
            df = pd.concat([df,new_row])
    else:
        new_row = pd.DataFrame({'old_id':row['id'],'coordinate1':[List[0]], 'coordinate2':[List[1]], 'coordinate1_lon':[List[0][0]], 'coordinate1_lat':[List[0][1]], 'coordinate2_lon':[List[1][0]], 'coordinate2_lat':[List[1][1]]})
        df = pd.concat([df,new_row])
road.apply(func, axis=1)
df
```

```python
df.drop_duplicates(['coordinate1_lon', 'coordinate1_lat', 'coordinate2_lon', 'coordinate2_lat'])
```

```python
df.to_csv('data/road1.csv', index=False) # list of edges
```

```python
road1 = pd.read_csv('data/road1.csv')
road1
```

```python
"""
    计算两个经纬度坐标之间的距离
    :param coord1, coord2: 第一个坐标，格式为 (经度1, 纬度1), (经度2, 纬度2)
    :return: 距离（单位：米）
"""
def calculate_distance(coord1, coord2):
    coord_1 = (coord1[1], coord1[0])
    coord_2 = (coord2[1], coord2[0])
    distance = geodesic(coord_1, coord_2).meters
    return distance

road1['dis']=0
def new_dis(row):
    c1 = eval(row['coordinate1'])
    c2 = eval(row['coordinate2'])
    row['dis'] = calculate_distance(c1, c2)
    return row
road1 = road1.apply(new_dis, axis=1)
```

```python
road1
```

```python
road1.to_csv('data/road1.csv', index=False)
```

```python
df2 = pd.DataFrame(columns=['coordinate'])
for i in range(len(road1)):
    row = road1.iloc[i,:]
    df2 = pd.concat([df2, pd.DataFrame({'coordinate':[row['coordinate1']]})])
    df2 = pd.concat([df2, pd.DataFrame({'coordinate':[row['coordinate2']]})])
df2
```

```python
df2 = df2.drop_duplicates()
df2
```
