```python
import pandas as pd
import numpy as np
```

```python
traj = pd.read_csv('../data/traj.csv')
traj
```

```python
to_delete = []
trajs = traj.groupby('traj_id')
for name, group in trajs:
    if len(group) < 15:
        to_delete.append(name)
print(len(to_delete))
traj = traj.drop(traj[traj['traj_id'].isin(to_delete)].index)
traj
```

```python
traj['x'] = traj.apply(lambda row:eval(row['coordinates'])[0],axis=1)
traj['y'] = traj.apply(lambda row:eval(row['coordinates'])[1],axis=1)
traj
```

```python
traj['x_diff'] = traj['x'].diff()
traj['y_diff'] = traj['y'].diff()
```

```python
traj['traj_id_changed'] = traj['traj_id'].ne(traj['traj_id'].shift())
traj = traj.loc[~traj['traj_id_changed'], :]
traj = traj.drop('traj_id_changed', axis=1)
traj
```

```python
traj['x_diff'].describe()
```

```python
traj['y_diff'].describe()
```

```python
traj['x_Digits'] = traj['x_diff'].apply(lambda x: [int(digit) for digit in "{:.6f}".format(x).split('.')[-1][:6]])
traj['y_Digits'] = traj['y_diff'].apply(lambda x: [int(digit) for digit in "{:.6f}".format(x).split('.')[-1][:6]])
traj
```

```python
traj['x_label0'] = np.sign(traj['x_diff']).astype(int)
traj['x_label1'] = traj['x_Digits'].apply(lambda x: x[0])
traj['x_label2'] = traj['x_Digits'].apply(lambda x: x[1])
traj['x_label3'] = traj['x_Digits'].apply(lambda x: x[2])
traj['x_label4'] = traj['x_Digits'].apply(lambda x: x[3])
traj['x_label5'] = traj['x_Digits'].apply(lambda x: x[4])
traj['x_label6'] = traj['x_Digits'].apply(lambda x: x[5])
traj['y_label0'] = np.sign(traj['y_diff']).astype(int)
traj['y_label1'] = traj['y_Digits'].apply(lambda x: x[0])
traj['y_label2'] = traj['y_Digits'].apply(lambda x: x[1])
traj['y_label3'] = traj['y_Digits'].apply(lambda x: x[2])
traj['y_label4'] = traj['y_Digits'].apply(lambda x: x[3])
traj['y_label5'] = traj['y_Digits'].apply(lambda x: x[4])
traj['y_label6'] = traj['y_Digits'].apply(lambda x: x[5])
traj
```

```python
traj = traj.drop(columns=['x_Digits', 'y_Digits', 'coordinates'])
traj
```
