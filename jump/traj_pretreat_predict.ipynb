```python
import pandas as pd
import numpy as np
```

```python
traj = pd.read_csv('../data/traj.csv')
traj
```

```python
to_delete = []
trajs = traj.groupby('traj_id')
for name, group in trajs:
    if len(group) < 15:
        to_delete.append(name)
print(len(to_delete))
traj = traj.drop(traj[traj['traj_id'].isin(to_delete)].index)
traj
```

```python
traj.groupby('traj_id').count().sort_values('id')
```

```python
traj['x'] = traj.apply(lambda row:eval(row['coordinates'])[0],axis=1)
traj['y'] = traj.apply(lambda row:eval(row['coordinates'])[1],axis=1)
traj
```

```python
(traj['x'] - 116).describe()
```

```python
traj['x'] = (traj['x'] - 116) * 100
```

```python
(traj['y'] - 39).describe()
```

```python
traj['y'] = (traj['y'] - 39) * 100
```

```python
traj['time'] = pd.to_datetime(traj['time']).astype('int64') / 10**9 * 100 % (1.38 * (10**11))
traj = traj.drop('coordinates',axis=1)
traj
```

```python
traj.to_csv('train_predict.csv',index=False)
```
