```python
import osmnx as ox
import networkx as nx
import pandas as pd
import matplotlib.pyplot as plt
import geopandas as gpd
from shapely.geometry import Point, LineString
```

```python
traj = pd.read_csv('../data/traj.csv')
traj
```

```python
trajs = traj.groupby('traj_id')
df = pd.DataFrame(columns=['traj_id', 'dis'])
for name, group in trajs:
    df = pd.concat([df,pd.DataFrame({'traj_id':[name], 'dis':[group.loc[group.index[-1],'current_dis']]})])
df
```

```python
df['dis'] = df['dis'].astype(float)
df = df[(df['dis'] < df['dis'].median() + 1) & (df['dis'] > df['dis'].median() - 1)]
chosen = df.sample(n=1)
chosen
```

```python
traj['traj_id'].astype(int)
df2 = traj.loc[traj['traj_id'] == chosen.loc[0, 'traj_id'],:]
df2
```

```python
base = []
def func(row):
    base.append(eval(row['coordinates']))
    return eval(row['coordinates'])
df2.apply(func, axis=1)
base
```

```python
from dtw import dtw
import numpy as np
base = np.array(base)
df3 = pd.DataFrame(columns=['traj_id', 'dis1',])

def euclidean_distance(x, y):
    return np.sqrt(np.sum((x - y) ** 2))


trajs = traj.groupby('traj_id')
for name, group in trajs:
    tmp = []
    def func(row):
        tmp.append(eval(row['coordinates']))
        return eval(row['coordinates'])
    group.apply(func, axis=1)
    tmp = np.array(tmp)
    distance = dtw(base, tmp, dist=euclidean_distance)
    
    df3 = pd.concat([df3,pd.DataFrame({'traj_id':[name], 'dis1':[distance[0]]})])
df3
```

```python
df3.to_csv('traj_mid_1.csv', index=False)
```

```python
# 将时间字符串转换为datetime对象
traj['time'] = pd.to_datetime(traj['time'])

# 提取时间段
traj['time_period'] = traj['time'].apply(lambda x: x.hour)
traj
```

```python
traj_time_first = traj[['traj_id','time_period']].drop_duplicates(subset=['traj_id'],keep='first')
traj_time_last = traj[['traj_id','time_period']].drop_duplicates(subset=['traj_id'],keep='last')
traj_time_last
```

```python
res = pd.merge(left=df3, right=traj_time_first, how='left', on='traj_id')
res = pd.merge(left=res, right=traj_time_last, how='left', on='traj_id')
res
```

```python
res['time_period_mid'] = (res['time_period_x'] + res['time_period_y'])/2
res
```

```python
res
```

```python
maxDis1 = res['dis1'].max()
minDis1 = res['dis1'].min()
def func1(row):
    return (row['dis1'] - minDis1)/(maxDis1 - minDis1)
def func2(row):
    return row['time_period_mid'] / 24      
res['dis1_res'] = res.apply(func1, axis=1)
res['dis2_res'] = res.apply(func2, axis=1)
res
```

```python
res.to_csv('traj_feature.csv',index=False)
```
